# print n natural no

```python
def naturaln(i):
    if i>0:
        print(i)
        i-=1
        naturaln(i)
naturaln(10)      
```

```python
def naturalno(n):
    if n>0:
        naturalno(n-1)
        print(n)
```

```python
naturalno(5)
```

# print n even no

```python
def Nevenno(n):
    if n>0:
        Nevenno(n-1)
        print(2*n)
```

```python
def even(i):
    if i<21:
        print(i)
        i+=2
        even(i)
even(2)
```

# reverse order even no

```python
def reverseven(n):
    if n>0:
        print(2*n)
        reverseven(n-1)
reverseven(5)
```

# sum of natural no

```python
def SumN(n):
    if n==1:
        return n
    return SumN(n-1)+n
```

```python
SumN(5)
```
